```python
from sklearn.datasets import fetch_openml
```

```python
dados = fetch_openml(data_id=44)
```

```python
type(dados)
```

```python
dados
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dados.data,columns=dados.feature_names)
```

```python
df.head(10)
```

```python
target_map={
    '0':'non-spam',
    '1':'spam'
}
```

```python
df['diagnostic'] = [target_map[target] for target in dados.target]
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Separating columns
x = df[['word_freq_make','word_freq_address','word_freq_all','word_freq_3d','word_freq_our', 'word_freq_over', 'word_freq_remove','word_freq_internet','word_freq_order','word_freq_mail', 'word_freq_people', 'word_freq_report', 'word_freq_business', 'word_freq_money', 'word_freq_telnet', 'char_freq_%3B', 'char_freq_%28', 'char_freq_%5B','char_freq_%21','char_freq_%24','char_freq_%23','capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total','word_freq_internet','word_freq_order','word_freq_mail','char_freq_%3B', 'char_freq_%28']]
y = df['diagnostic']

# Separating lines
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler

escalonador = MinMaxScaler()

escalonador.fit(x_train)

x_train_escalonados = escalonador.transform(x_train)
x_test_escalonados = escalonador.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train_escalonados, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = knn.predict(x_test_escalonados)

print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples',
            xticklabels=['non-spam', 'spam'], yticklabels=['non-spam', 'spam'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
#True Positives (TP) = 261: True spam messages that were correctly classified as spam.

#False Negatives (FN) = 102: Spam messages that were wrongly classified as non-spam.

#False Positives (FP) = 31: Legitimate (non-spam) messages that were wrongly classified as spam.

#True Negatives (TN) = 527: Non-spam messages that were correctly classified.
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=10)
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)

print(classification_report(y_test, y_pred_rf))
```

```python
import matplotlib.pyplot as plt
# Calculate the confusion matrix for RandomForest
cm_rf = confusion_matrix(y_test, y_pred_rf)

# Plot the confusion matrix for RandomForest
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Purples',
            xticklabels=['non-spam', 'spam'], yticklabels=['non-spam', 'spam'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - RandomForest')
plt.show()
```

```python
#True Positives (TP) = 317: True spam messages that were correctly classified as spam.

#False Negatives (FN) = 46: Spam messages that were wrongly classified as non-spam.

#False Positives (FP) = 22: Legitimate (non-spam) messages that were wrongly classified as spam.

#True Negatives (TN) = 536: Non-spam messages that were correctly classified.
```

```python
# Correlation Analysis
# We only take the columns of words
word_features = [col for col in df.columns if col.startswith('word_freq_')]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnostic_numeric'] = le.fit_transform(df['diagnostic'])

# Correlation with the variable 'diagnostic_numeric'
correlation = df[word_features].corrwith(df['diagnostic_numeric'])
correlation = correlation.sort_values(ascending=False)

# View the top 10 words most correlated with spam
top_words = correlation.head(10)
print(top_words)

# Graphic
top_words.plot(kind='barh', color='red')
plt.title('Top words most correlated with SPAM')
plt.xlabel('Correlation with Spam')
plt.gca().invert_yaxis()
plt.show()
```
